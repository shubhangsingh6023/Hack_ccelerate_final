### Cleaning Data and then Organising as per our requirement

```python
#I need to define another variable Car_Name which would be maker + model. So, get a better idea of any particular car
df['Car_Name'] = df.Make + ' ' + df.Model
df.sample(5)
```

```python
#Since there are around 140 columns (features) but depending on the indian market system and customers and reviews
#from my family members and other local people, I sorted some specific columns (features) and thus used that dataset
#for my furthur approach. Also working on 140 columns was bit complicated and the decison made would be ambigious.
necessary_features = ['Make','Model','Car_Name','Variant','Body_Type','Fuel_Type','Fuel_System','Type','Drivetrain',
                      'Ex-Showroom_Price','Displacement','Cylinders','ARAI_Certified_Mileage','Power','Torque',
                      'Fuel_Tank_Capacity','Height','Length','Width','Doors','Seating_Capacity','Wheelbase',
                      'Number_of_Airbags']
#For doing the data analysis on the data set we need some more numerical values such as tank capacity, Ex-showroom 
#price and all other need to be converted to float type
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].apply(str).str.replace( 'Rs. ' , '' ,regex = False)#replaces Rs 
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].apply(str).str.replace( ',' , '' ,regex = False)#replaces , 
df['Ex-Showroom_Price'] = df['Ex-Showroom_Price'].apply(str).astype(int)# coverts to integer
df = df[necessary_features] #defining data for only necessary features
df = df[~df.ARAI_Certified_Mileage.isnull()] #null values in arai certified mileage is removed
df = df[~df.Make.isnull()] #null values in make (car company) is removed
df = df[~df.Width.isnull()]#null values in width is removed
df = df[~df.Cylinders.isnull()]#null values in number of cylinders is removed
df = df[~df.Wheelbase.isnull()]#null values in wheelbase is removed
df = df[~df['Fuel_Tank_Capacity'].isnull()]#null values in Tank Capacity is removed
df = df[~df['Seating_Capacity'].isnull()]#null values in seating capacity is removed
df = df[~df['Torque'].isnull()]#null values in torque is removed
df['Height'] = df['Height'].apply(str).str.replace(' mm','',regex=False).astype(float) #replaces mm and converts it to float in Height column
df['Length'] = df['Length'].apply(str).str.replace(' mm','',regex=False).astype(float) #replaces mm and converts it to float in Length column
df['Width'] = df['Width'].apply(str).str.replace(' mm','',regex=False).astype(float)  #replaces mm and coverts it to float in Width and Wheelbase column
df['Wheelbase'] = df['Wheelbase'].apply(str).str.replace(' mm','',regex=False).astype(float)#apply(str) is used for sorting some error
df['Fuel_Tank_Capacity'] = df['Fuel_Tank_Capacity'].apply(str).str.replace(' litres','',regex=False).astype(float) #replaces litres and coverts to float in Tank Capacity column

#We still some more calculative terms (on which we can use maths on) for our EDA so I defined some more columns
df['Displacement'] = df['Displacement'].apply(str).str.replace(' cc','',regex=False) #removing cc 
df.loc[df.ARAI_Certified_Mileage == '9.8-10.0 km/litre','ARAI_Certified_Mileage'] = '10'# for 9.8-10 considering as 10
df.loc[df.ARAI_Certified_Mileage == '10kmpl km/litre','ARAI_Certified_Mileage'] = '10'
df['ARAI_Certified_Mileage'] = df['ARAI_Certified_Mileage'].apply(str).str.replace(' km/litre','',regex=False).astype(float) #replacing str terms and then converting to float
df.Number_of_Airbags.fillna(0,inplace= True) #filling na with 0
df['GST_price'] = df['Ex-Showroom_Price'] * 0.014  #Indian customers usually need on road price 
#rather than ex showrrom price a report claims so I calculated aprrox on road price by multiplying with 14% GST
df.GST_price = df.GST_price.astype(int) #coverting it to int type
df['On_road-price'] = df['GST_price'] + df['Ex-Showroom_Price'] #calculating on road price from GST + ex showroom price
df.drop(columns = 'Ex-Showroom_Price', inplace= True )
df.drop(columns = 'GST_price', inplace= True )
df.drop(columns = 'Fuel_System', inplace= True )
HP = df.Power.str.extract(r'(\d{1,4}).*').astype(int) * 0.98632
HP = HP.apply(lambda x: round(x,2)) #calculating power
TQ = df.Torque.str.extract(r'(\d{1,4}).*').astype(int)
TQ = TQ.apply(lambda x: round(x,2)) #calculating torque
df.Torque = TQ
df.Power = HP
df.Doors = df.Doors.astype(int) #converting to integer
df.Seating_Capacity = df.Seating_Capacity.astype(int)
df.Number_of_Airbags = df.Number_of_Airbags.astype(int)
df.Displacement = df.Displacement.astype(int)
df.Cylinders = df.Cylinders.astype(int)
#finally these columns are required to me in the cleaned dataset
df.columns = ['Manufacturer', 'Model', 'Car_Name', 'Variant', 'Body_Type', 'Fuel_Type','Gear_Type', 'Drivetrain', 'Displacement', 'Cylinders',
              'Mileage', 'Power', 'Torque', 'Fuel_tank','Height', 'Length', 'Width', 'Doors', 'Seats', 'Wheelbase','Airbags', 'On_road_price']
df = df.dropna() #if some na values are still left then drop those
df.sample(4)
```
